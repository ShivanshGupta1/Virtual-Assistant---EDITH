```python
import pyttsx3 # text to speech
import speech_recognition as sr #speech recognition

engine = pyttsx3.init()
engine.setProperty('rate',178)



def speak(words):
    engine.say(words)
    engine.runAndWait()

def listen():
    r = sr.Recognizer()
    with sr.Microphone() as mic: #with (sr.Microphone) -> open
        print('Listening from Microphone...')
        r.adjust_for_ambient_noise(mic)
        r.pause_threshold = 1
        audio = r.listen(mic)
   
    try:
        print("Recognizing Speech")
        query = r.recognize_google(audio,language='en-in')
       
    except Exception as e: #Exception is a type of error
        print('Error: ',e)
        print("Unable to recognize this query! Kindly RUN AGAIN!")
        return None
   
    return query

text = listen()
speak(text)

#with is a keyword to open files/running things like - with open
#but the sr already has open, so we just use with!

speak('Hello World!')

## if does not work use - conda install -c anaconda pyaudio
```

```python
import pyttsx3 # text to speech
import speech_recognition as sr #speech recognition
import datetime as dt
import pyjokes as pj
import requests, json
import os, wolframalpha

engine = pyttsx3.init()
engine.setProperty('rate',200)


def speak(words):
    engine.say(words)
    engine.runAndWait()

def listen():
    r = sr.Recognizer()
    with sr.Microphone() as mic: #with (sr.Microphone) -> open
        print('Listening from Microphone...')
        r.adjust_for_ambient_noise(mic)
        r.pause_threshold = 1
        audio = r.listen(mic)
   
    try:
        print("Recognizing Speech")
        query = r.recognize_google(audio,language='en-in')
       
    except Exception as e: #Exception is a type of error
        print('Error: ',e)
        print("Unable to recognize this query! Kindly RUN AGAIN!")
        return None
   
    return query


speak("Hi I am edith, an A I program developed by Tony Stark! How may I help you today?")
while True:
    command = listen()
    if command != None:
        command = command.lower()
        if "your name" in command:
            speak("My name is edith, which stands for even dead I am the hero!")
        elif "date today" in command or "today's date" in command:
            myDate = dt.date.today()
            speak(myDate.strftime("%A %d %B %Y")) #str stands for string as this converts it into a string
        elif "time" in command:
            myTime = dt.datetime.now()
            speak(myTime.strftime("%I:%M:%S:%p"))
        elif "weather" in command:     
            base_url = "http://api.openweathermap.org/data/2.5/weather?q={}&units=metric&appid=c6c30d38cf009bf578bec2302aa23248"
            speak("Please tell me the City Name!")
            print("Waiting for city name...")
            city_name = listen()
            complete_url = base_url.format(city_name)
            response = requests.get(complete_url)
            x = response.json()
            if x["cod"]  != "404":
                y = x["main"]
                current_temperature = y["temp"]
                current_pressure = y["pressure"]
                current_humidiy = y["humidity"]
                weather = x["weather"]
                weather_description = weather[0]["description"]
                speak(" Temperature in celsius unit is "+str(current_temperature)+", atmospheric pressure in hPa unit is"+str(current_pressure) +", humidity in percentage is " +str(current_humidiy) +"and the description mentioned is " +str(weather_description))         
            else:
                speak(" City Not Found ")
        elif "write a note" in command:
            myFile = open("vaNotes.txt","w")
            speak("What should I write Sir?")
            note = listen();
            speak("Sir, should I include the date and time too?")
            dateTime = listen()
            if dateTime!=None:
                if "yes" in dateTime:
                    dateToday = dt.date.today()
                    timeToday = dt.datetime.now()
                    myFile.write("Date: "+ dateToday.strftime("%A %d %B %Y")+"\n")
                    myFile.write("Time: "+timeToday.strftime("%I:%M:%S:%p")+"\n")
                    myFile.write(note)
                else:
                    myFile.write(note)
            myFile.close()
        elif "open" in command:
            speak("Please tell the application name Sir!")
            nameApp = listen()
            nameApp = nameApp.title()
            filePath = f"open /Applications/{nameApp}.app"
            os.system(filePath)
        elif "read note" in command:
            myFile = open("vaNotes.txt","r")
            fileInfo = myFile.readlines()
            for element in fileInfo:
                speak(element)
            speak("That was all in the file Sir!")
            myFile.close()
        elif "calculate" in command:
            command=command.replace("calculate","")
            apiId = "R3LJXJ-7QR84GVE2H"
            client = wolframalpha.Client(apiId)
            response = client.query(command)
            answer = next(response.results).text
            speak(answer)
        elif "joke" in command or "jokes" in command:
            speak(pj.get_joke())
        elif "exit" in command or "bye" in command:
            speak("Thank you for using me! Hope we speak again")
            break
```
